```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

## **Basic Graph**

```python
x = [0,1,2,3,4]
y = [0,2,4,6,8]

plt.figure(figsize = (8, 5), dpi = 100)

#plt.plot(x, y, label = '2x', color = 'red',linestyle='dashed' , linewidth = 2, marker = '.', markersize = 10)

plt.plot(x, y, 'r^--' , label = '2x') # Shorthand notation = '[Color][Marker][Dashline]'

#Line 2

x2 = np.arange(0, 4.5, 0.5)
plt.plot(x2[:5], x2[:5]**2, label='x^2')
plt.plot(x2[4:], x2[4:]**2, 'b--')

plt.title('Our first graph', fontdict = {'fontname': 'DejaVu Sans', 'fontsize':20})
plt.xlabel("X-axis")
plt.ylabel("Y-axis")

plt.xticks([0,1,2,3,4,])
#plt.yticks([0,2,4,6,8,10,])

plt.legend()

plt.savefig("mygraph.png", dpi =300)

plt.show()
```

## **Bar Chart**

```python
values = [1, 4, 2]
labels = ['A', 'B', 'C']

plt.figure(figsize = (5, 3), dpi = 100)

bars = plt.bar(labels, values)
patterns = ['/', 'O', '*']
for bar in bars:
    bar.set_hatch(patterns.pop(0))

plt.show()

plt.savefig("mybarchart", dpi = 300)
```

### **Line Graph**

```python
gas = pd.read_csv('https://raw.githubusercontent.com/KeithGalli/matplotlib_tutorial/master/gas_prices.csv')
gas.head(5)
```

```python
from matplotlib.font_manager import FontProperties

fontP = FontProperties()
fontP.set_size('small')
```

```python
plt.figure(figsize = (8, 5))

# plt.plot(gas.Year, gas.USA, 'b.-' , label = "United States")
# plt.plot(gas.Year, gas.Canada, 'r.-' ,label = "Canada")
# plt.plot(gas.Year, gas['South Korea'], 'g.-', label = "South Korea")
# plt.plot(gas.Year, gas.Australia, 'y.-' ,label = "Australia")

for country in gas:
    if country != 'Year':
        plt.plot(gas.Year, gas[country], marker='.', label = country)

plt.title('Gas Prices over Time (in USD)', fontdict = {'fontweight':'bold', 'fontsize': 18})
plt.xlabel("Year")
plt.ylabel("US Dollars")

plt.xticks(gas.Year[::3])

plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left', prop=fontP)

#plt.savefig("Gas_price_figure.png", dpi = 300, bbox_inches="tight")

plt.show()
```

## **Load FIFA Data**

```python
fifa = pd.read_csv('https://raw.githubusercontent.com/KeithGalli/matplotlib_tutorial/master/fifa_data.csv')

fifa.head(5)
```

## **Histogram**

```python
bins = np.arange(40,110,10)
print(bins)

plt.figure(figsize=(8,5))

plt.hist(fifa.Overall, bins = bins, color = "#60f798", edgecolor='black', linewidth=0.5)

plt.ylabel('Number of Players')
plt.xlabel('Skill Level')
plt.title('Distribution of Player Skills in FIFA 2018', fontdict = {'fontweight':'bold', 'fontsize': 14})

plt.xticks(bins)

#plt.savefig("histogram.png", dpi =300)

plt.show()
```

## **Pie Chart**

```python
fifa.columns
```

```python
left = fifa.loc[fifa['Preferred Foot'] == 'Left'].count()[0]
right = fifa.loc[fifa['Preferred Foot'] == 'Right'].count()[0]

plt.figure(figsize = (8, 5))

labels = ['Left', 'Right']
color = ['#32a852', '#8ec6ed']

plt.pie([left, right], labels = labels, colors=color, autopct = '%.2f%%')

plt.title('Foot Preference of FIFA Players', fontdict = {'fontsize': 14})

#plt.savefig("FootPreference.png", dpi = 300)

plt.show()
```

## **Pie Chart#2**

```python
fifa.Weight = [int(x.strip('lbs')) if type(x)==str else x for x in fifa.Weight]

plt.figure(figsize=(8,5), dpi=100)

plt.style.use('seaborn-darkgrid')

light = fifa.loc[fifa.Weight < 125].count()[0]
light_medium = fifa[(fifa.Weight >= 125) & (fifa.Weight < 150)].count()[0]
medium = fifa[(fifa.Weight >= 150) & (fifa.Weight < 175)].count()[0]
medium_heavy = fifa[(fifa.Weight >= 175) & (fifa.Weight < 200)].count()[0]
heavy = fifa[fifa.Weight >= 200].count()[0]

weights = [light,light_medium, medium, medium_heavy, heavy]
label = ['under 125', '125-150', '150-175', '175-200', 'over 200']
explode = (.4,.1,0,0,.4)

plt.pie(weights, labels = label, autopct = '%.2f%%', pctdistance=0.8, explode=explode)

plt.title('Weight of Professional Football Players (lbs)')

plt.savefig("WeightDistribution.png", dpi =300)

plt.show()
```

## **Box and Whiskers Plot**

```python
plt.figure(figsize = (5,7), dpi = 100)

plt.style.use('default')

barcelona = fifa.loc[fifa.Club == 'FC Barcelona']['Overall']
madrid = fifa.loc[fifa.Club == 'Real Madrid']['Overall']
revs = fifa.loc[fifa.Club == "New England Revolution"]['Overall']

label = ['FC Barcelona', 'Real Madrid', 'NE Revolution']

plt.boxplot([barcelona, madrid, revs], labels = label, patch_artist=True, medianprops={'linewidth': 2})

plt.title('Professional Football Team Comparison')
plt.ylabel('FIFA Overall Rating')

plt.savefig("TeamComperison.png", dp= 300)

plt.show()
```
